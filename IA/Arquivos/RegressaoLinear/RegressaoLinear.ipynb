```python
from sklearn.datasets import make_regression #função que gera dados aleatórios que possam ser relacionados por regressão linear
import matplotlib.pyplot as plt
x,y = make_regression(n_samples=200,n_features=1,noise=30)#n_samples é a quantidade de dados gerados, 
#noise é o nível de discrepância entre os dados e as dimensões do gráfico serão n_features + 1
plt.scatter(x,y)#define que x se relacionará a y
plt.show()#exibe um gráfico com os valores gerados
```

```python
from sklearn.linear_model import LinearRegression #função que vai fazer a regressão linear com os dados imformados
modelo = LinearRegression() #aloca a função dentro da variável modelo
```

```python
modelo.fit(x,y) #passa os valores como parâmetro
```

```python
modelo.intercept_ #exibe o coeficiente linear da reta "b"
```

```python
modelo.coef_ #exibe o coeficiente angular da reta "m"
```

```python
import numpy as np #biblioteca matemática
plt.scatter(x,y)
xreg = np.arange(-5,4,1)# cria uma reta onde o x varia de -3 até 4, com interval de 1 em 1 entre os valores 
#(observar variação) de x no gráfico
plt.plot(xreg,modelo.coef_*xreg-modelo.intercept_,color='red')#implementa uma fórmula em uma variável, no caso: mx-b
# além disso gera uma reta para representar o resultado
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_treino, x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.30) #divide os dados entre teste e treino de acordo com
# a porcentagem informada, no caso 0.30 = 30%
```

```python
modelo.fit(x_treino,y_treino)
```

```python
resultado = modelo.score(x_teste,y_teste) #retorna o coeficiente R2 do modelo, ou seja, testa o modelo com os dados de teste 
#e calcula a taxa de precisão, quanto mais próximo de 1 melhor
```

```python
print(resultado)
```
