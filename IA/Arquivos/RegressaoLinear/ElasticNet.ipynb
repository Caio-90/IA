```python
# para isso definiremos valores para uma constante(C) que representará o peso das fuções L1 e L2, esse peso deve ser um valor 
# entre 0 e 1, o peso da função L1 será a própria constante, já o da L2 será 1 - constante, dado pela fórmula: 
# ERRO + C(L1) + (1-C)(L2)
# desse forma uma constante com valor 0.5 significa 50% de peso para cada, 1 significa 100% de peso para a L1, e por aí vai
```

```python
from sklearn.linear_model import ElasticNet
import pandas as pd 
pd.set_option('display.max_columns',21)
dados = pd.read_csv('C:/Users/Cliente/IA/basesDeDados/kc_house_data.csv')
dados.head()
#importar a base de dados
```

```python
from sklearn.model_selection import train_test_split
dados.drop('id',axis=1,inplace=True)
dados.drop('date',axis=1,inplace=True)
dados.drop('zipcode',axis=1,inplace=True)
dados.drop('lat',axis=1,inplace=True)
dados.drop('long',axis=1,inplace=True)
```

```python
y = dados['price'] 
x = dados.drop('price',axis=1) 
x_treino, x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.30)
```

```python
modelo = ElasticNet(l1_ratio = 0.9,tol = 0.001,max_iter = 5000) #l1 ratio é a constante que define os pesos
modelo.fit(x_treino,y_treino)
resultado = modelo.score(x_teste,y_teste)
print (resultado)
```
