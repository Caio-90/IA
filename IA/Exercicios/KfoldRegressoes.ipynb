```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',21)
dados = pd.read_csv('C:/Users/Cliente/IA/basesDeDados/kc_house_data.csv')
dados.head()
# utilizarei a maior e mais complexa base de dados que possuo para averiguar a eficácia do modelo
```

```python
dados.drop('id',axis=1,inplace=True)
dados.drop('date',axis=1,inplace=True)
dados.drop('zipcode',axis=1,inplace=True)
dados.drop('lat',axis=1,inplace=True)
dados.drop('long',axis=1,inplace=True)
# excluir colunas irrelevantes e não numéricas
```

```python
y = dados['price'] 
x = dados.drop('price',axis=1) 
```

```python
def modelos(x,y):
    from sklearn.linear_model import Lasso
    from sklearn.linear_model import Ridge
    from sklearn.linear_model import LinearRegression
    from sklearn.linear_model import ElasticNet
    from sklearn.model_selection import cross_val_score, KFold
    
    reg = LinearRegression()
    ridge = Ridge()
    lasso = Lasso()
    elastic = ElasticNet()
    
    kfold  = KFold(n_splits = 5)
    
    Rreg = cross_val_score(reg,x,y,cv = kfold)
    RFreg = max(Rreg)
    Rridge = cross_val_score(ridge,x,y,cv = kfold)
    RFridge = max(Rridge)
    Rlasso = cross_val_score(lasso,x,y,cv = kfold)
    RFlasso = max(Rlasso)
    Relastic = cross_val_score(elastic,x,y,cv = kfold)
    RFelastic = max(Relastic)
    
    print('Regressão linear:',RFreg,'Ridge:',RFridge,'Lasso:',RFlasso,'Elastic:',RFelastic)
```

```python
modelos(x,y) # os resultado melhoraram em torno de 0.4% e 1.2%
```
